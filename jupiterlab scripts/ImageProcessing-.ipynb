```python
full_dir = "E:/User Document/dataset" 
p = os.listdir(full_dir+'/PhosphorusLow')
n = os.listdir(full_dir+'/NitrogenLow')
k = os.listdir(full_dir+'/PotassiumLow')
ph = os.listdir(full_dir+'/PhosphorusHigh')
nh = os.listdir(full_dir+'/NitrogenHigh')
kh = os.listdir(full_dir+'/PotassiumHigh')


nlist=[]
plist=[]
klist=[]
nhlist=[]
phlist=[]
khlist=[]

for ne in n  :
    nlist.append(os.path.join(full_dir+"/NitrogenLow",ne))

for ne in p  :
    plist.append(os.path.join(full_dir+"/PhosphorusLow",ne))
for ne in k  :
    klist.append(os.path.join(full_dir+"/PotassiumLow",ne))
for ne in nh :
    nhlist.append(os.path.join(full_dir+"/NitrogenHigh",ne))
for ne in ph :
    phlist.append(os.path.join(full_dir+"/PhosphorusHigh",ne))
for ne in kh :
    khlist.append(os.path.join(full_dir+"/PotassiumHigh",ne))
    

```

```python
scale_percent = 220

def preocessImage(im1):
    image = Image.fromarray(im1)
    img = image.resize((180, 540))
    image_np = np.array(img)
    
    im1 = image_np
    leaf_hsv = rgb2hsv(im1)
    
    lower_mask = leaf_hsv[:,:,0] > 0.0 
    upper_mask = leaf_hsv[:,:,0] < 0.2 
    saturation_mask = leaf_hsv[:,:,1] > 0
    mask2 =upper_mask*lower_mask*saturation_mask
    red = im1[:,:,0]*mask2
    green = im1[:,:,1]*mask2
    blue = im1[:,:,2]*mask2
    leaf_masked_symptom = np.dstack((red,green,blue))
       
    lower_mask = leaf_hsv[:,:,0] > 0.2 
    upper_mask = leaf_hsv[:,:,0] < 0.3
    saturation_mask = leaf_hsv[:,:,1] > 0.4 
    
    mask1 = upper_mask*lower_mask*saturation_mask
    red = im1[:,:,0]*mask1
    green = im1[:,:,1]*mask1
    blue = im1[:,:,2]*mask1
    leaf_masked_other = np.dstack((red,green,blue))
   
    black_pixels = np.where(
            (leaf_masked_other[:, :, 0] == 0) & 
            (leaf_masked_other[:, :, 1] == 0) & 
            (leaf_masked_other[:, :, 2] == 0)
            )
    leaf_masked_other[black_pixels] = [0, 0, 0]
    return cv2.fastNlMeansDenoisingColored(leaf_masked_symptom+leaf_masked_other ,None,10,10,7,21)
    
name = 0;
im1 = imageio.imread(nhlist[25])
x=preocessImage(im1);
imageio.imsave("E:/User Document/processed2/"+str(name)+".jpg", x)
im1 = imageio.imread(phlist[25])
x=preocessImage(im1);
name = 1
imageio.imsave("E:/User Document/processed2/"+str(name)+".jpg", x)
im1 = imageio.imread(khlist[25])
x=preocessImage(im1);
name = 2
imageio.imsave("E:/User Document/processed2/"+str(name)+".jpg", x)

    
```
